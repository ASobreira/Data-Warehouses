```python
# Create new column with State Abbreviations
customers_USA['State_abb'] = customers_USA['State'].replace(us_state_to_abbrev)

# Create Choropleth object
choro = go.Choropleth(
    locations= customers_USA['State_abb'].value_counts().index, # Spatial coordinates
    z = customers_USA['State_abb'].value_counts(), # costumers_USA to be color-coded
    locationmode = 'USA-states',
    colorscale = 'teal', zmin = 1, zmax = 1000,
    text = customers_USA['State_abb']
)

# Create Figure object
fig = go.Figure(data=choro, layout=go.Layout(
    font = dict(
            size = 14
            ),    
    title={
       # 'text': "Number of Customers by State Map",
        'y':0.95,
        'x':0.5
        },
    geo_scope='usa' # limit map scope to USA
))

# Display Figure
fig.show()
```

##### States

```python
## CREATE NEW TABLE
# group by category
profit_per_states = GDP_USA[['State', 'GDP (Chained $)']].groupby('State').sum().head(25)

# sort values by Sales
profit_per_states.sort_values(by='GDP (Chained $)', ascending=False, inplace=True)


# bar chart
fig = px.bar(x = profit_per_states.index,
             y = profit_per_states['GDP (Chained $)'],
             color = profit_per_states.index,
             template = 'simple_white', color_discrete_sequence = colors)

#Set Title and x/y axis labels
fig.update_layout(
    xaxis_title="State",
    yaxis_title="GDP",
    showlegend = False,
    font = dict(
            size = 14
            ),    
    title={
        #'text': "GDP by State(25)",
        'y':0.95,
        'x':0.5
        }
    )

fig.show()
```

##### Region

```python
## CREATE NEW TABLE
# group by category
profit_per_states = GDP_USA[['Region', 'GDP (Chained $)']].groupby('Region').sum()

# sort values by Sales
profit_per_states.sort_values(by='GDP (Chained $)', ascending=False, inplace=True)

# bar chart
fig = px.bar(x = profit_per_states.index,
             y = profit_per_states['GDP (Chained $)'],
             color = profit_per_states.index,
             template = 'simple_white', color_discrete_sequence = colors)

#Set Title and x/y axis labels
fig.update_layout(
    xaxis_title="Region",
    yaxis_title="GDP",
    showlegend = False,
    font = dict(
            size = 14
            ),    
    title={
        #'text': "GDP by Region(25)",
        'y':0.95,
        'x':0.5
        }
    )

fig.show()
```
